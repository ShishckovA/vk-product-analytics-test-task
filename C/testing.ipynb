```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("task_for_a_prod_analytics_trainee_part_c.csv")
df.head()
```

```python
df["device_brand"].value_counts()
```

```python
df["is_male"] = (df["user_sex"] == "male").astype(int)
df = df.drop(["user_sex", "hashed_user_id", "device_brand", "device_model"], axis=1)
display(df.head())
display(df.shape)
```

```python
# from sklearn.preprocessing import StandardScaler
X = df.values
```

```python
from sklearn.cluster import *

birch = Birch(threshold=0.65)
cluster = birch.fit(X)
```

```python
df["cluster"] = cluster
(cluster == 0).sum()
```

```python
df.groupby("cluster").mean()
```
